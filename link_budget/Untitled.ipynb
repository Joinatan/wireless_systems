```python
import scipy as sp
import numpy as np
from matplotlib import pyplot as plt
signal_strengths = [-43, -48, -73, -70, -82, -75, -88, -109]
meters = [5, 10, 20, 30, 40, 50, 60, 70]
sp.stats.linregress(meters, signal_strengths)

def fu(x):
    return -0.838 * x - 43.6

dist = range(0, 100)

signalList = []
for i in dist:
    signalList.append(fu(i))
plt.plot(dist, signalList)
```

```python
def mean(dBList):
    lista = []
    for el in dBList:
        lista.append(10**(el/10))
        
    return 10 * np.log10((sum(lista)/len(lista)))
```

```python
enMeter= [-49, -44, -42, -45, -44]
tvaMeter = [-52, -56, -55, -60, -53]
fyraMeter = [-53, -57, -58, -60, -57]
attaMeter = [-62, -59, -59, -56, -53]
sextonMeter = [-63, -65, -58, -62, -60]
trettiotvaMeter = [-70, -72, -79, -80, -81]
avstand = [enMeter, tvaMeter, fyraMeter, attaMeter, sextonMeter, trettiotvaMeter]

mean_list = []
for el in avstand:
    mean_list.append(mean(el))

reversed_list = list(reversed(mean_list))
print(reversed_list)
print(mean_list)
plt.plot([2**x for x in range(0, 6)], mean_list)
plt.plot([0, 16, 26, 28, 30, 32], reversed_list)
```

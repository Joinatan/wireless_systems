```python
import numpy as np
from matplotlib import pyplot as plt

def todB(num):
    return 10 * np.log10(num)
def toLin(num):
    return 10**(num/10)
```

```python
def rayleighCDF(x):
    return 1-np.exp(np.e**-x)
```

```python
def ricianModel(amp, freq, n=1000):
    directSignal = amp
    reflectedSignals = 0
    for i in range(0, n):
        reflectedSignals += amp/2 * np.random.normal() 
        
    return (directSignal + reflectedSignals)
```

```python
signalList = []

for i in range(0, 1000):
    signalList.append(ricianModel(50, 18e6))
    
plt.plot(range(0, 1000), signalList)
```

```python
# availbility fade margin
def availbility(margin):
    return np.exp(-1/margin)

print('availbility 19.08dB', availbility(19.08))
print('availbility 31.12dB', availbility(31.12))
print('availbility 38.17dB', availbility(38.17))
```
